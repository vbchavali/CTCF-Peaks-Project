```python
import pandas as pd 
import csv

df_Nakahashi_upstream = pd.read_csv('df_Nakahashi_upstream.bed', sep = '\t')
df_Nakahashi_upstream.columns.values[0] = "chrom"
display(df_Nakahashi_upstream.columns)
display(df_Nakahashi_upstream)
df_scan_up = pd.read_csv('~/N_up.bed', sep='\t', header = 5)
#df_scan_down = pd.read_csv('~/N_down.bed', sep='\t', header = 4)

df_scan_up = df_scan_up.columns.to_frame().T.append(df_scan_up, ignore_index=True)
df_scan_up.columns = range(len(df_scan_up.columns))
df_scan_up.columns = ["chrom", "start", "end", "motif", "score", "strand"]
df_scan_up = df_scan_up.drop(columns=['motif', 'score', 'strand'])
display(df_scan_up.columns)
display(df_scan_up)
```

```python
import bioframe as bf

overlapping_intervals = bf.overlap(df_Nakahashi_upstream, df_scan_up, how = 'inner', suffixes = ('_1', '_2'))
display (overlapping_intervals)
```
